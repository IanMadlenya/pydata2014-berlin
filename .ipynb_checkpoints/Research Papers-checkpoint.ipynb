#### Collect the category links

```python
from bs4 import BeautifulSoup
import urllib2

url = "http://research.google.com/pubs/papers.html" # Base URL
raw = urllib2.urlopen(url).read() # Download page
soup = BeautifulSoup(raw) # Read HTML
links = [l.get("href") for l in soup.find_all("a")] # Identify links
links = ["http://research.google.com" + l for l in links if "/pubs/" in l] # Only publication links
links = [l for l in links if l != url] # Exclude self-referential links

links[:10]
```

### Get the papers

```python
title = list()
url = list()
abstract = list()

for l in links:
    raw = urllib2.urlopen(l).read()
    soup = BeautifulSoup(raw)
    title = title + [t.string.strip() for t in soup.find_all("p", class_="pub-title")] # Title in pub-title
    abstract = abstract + [a["href"] for a in soup.find_all("a", class_="abstract-icon tooltip")]

print title[:10]
```

### Clean the titles

```python
import re
title = [re.sub(' +',' ',t) for t in title]
title = [t.replace('\n','') for t in title]
print title[:10]
```

```python
papers = zip(title, abstract)
```

#### Liste mit den verwendeten Wörtern erstellen

```python
words = [ w.lower() for t in title for w in t.split() ]

words[:10]
```

#### Häufig verwendete Begriffe (Stopwords) entfernen

```python
import nltk
# Folgende Zeile entkommentieren, falls Korpus stopwords nicht gefunden wird -> Download der Stopwords
# nltk.download() # Öffnet Download-Manager

stop = nltk.corpus.stopwords.words('english')
print stop[:10]
words_without_stopwords = [i for i in words if i not in stop]
print
words_without_stopwords[:10]
```

#### Häufigkeitsverteilung

```python
from collections import Counter

print "Alle Wörter"
for item in [words]:
    c = Counter(item)
    print c.most_common()[:10] # top 10
    
print "Ohne Stpowords"
for item in [words_without_stopwords]:
    c = Counter(item)
    print c.most_common()[:10] # top 10
```

#### Tabelle mit Pretty Table erstellen

```python
from prettytable import PrettyTable

for label, data in (('Alle Wörter', words), 
                    ('Ohne Stopwörter', words_without_stopwords)):
    pt = PrettyTable(field_names=[label, 'Häufigkeit']) 
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[:100] ]
    pt.align[label], pt.align['Häufigkeit'] = 'l', 'r' # Set column alignment
    print pt
```

#### Kollokationen finden (gemeinsam verwendete Begriffe)

```python
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()

raw = " ".join(title)
tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)

text.collocations()
```

#### Reduktion der Wörter auf die Grundformen

```python
from prettytable import PrettyTable

porter = nltk.PorterStemmer()
text_stemmed = [porter.stem(t).lower() for t in tokens if t not in stop]

pt = PrettyTable(field_names=['Wortstamm', 'Häufigkeit']) 
c = Counter(text_stemmed)
[ pt.add_row(kv) for kv in c.most_common()[:20] ]
pt.align['Wortstamm'], pt.align['Häufigkeit'] = 'l', 'r' # Set column alignment
print pt
```

#### Lemmatisierung

```python
from prettytable import PrettyTable

# Lemmatisierung benötigt den nltk Corpus "wordnet"
# Installieren mit nltk.download() in der ipython-Shell

wnl = nltk.WordNetLemmatizer()
tokens_lower = [w.lower() for w in tokens]
text_lemmata = [wnl.lemmatize(t) for t in tokens_lower if t not in stop]

# Satzzeichen entfernen
text_lemmata = filter(lambda word: word not in ',-:', text_lemmata)

pt = PrettyTable(field_names=['Lemma', 'Häufigkeit']) 
c = Counter(text_lemmata)
[ pt.add_row(kv) for kv in c.most_common()[:250] ]
pt.align['Wortstamm'], pt.align['Häufigkeit'] = 'l', 'r' # Set column alignment
print pt
```

#### Download der Abstracts

```python
import re

print len(abstract)
abstracts = list()
title = list()
year = list()
author = list()
journal = list()

for l in abstract:
    raw = urllib2.urlopen("http://research.google.com" + l).read() # Seite öffnen und herunterladen
    soup = BeautifulSoup(raw) # Seite in HTML-Parser einlesen
    abstracts = abstracts + [t.text.strip().lower() for t in (td.find('div') for td in soup.findAll('div', class_="maia-col-8")) if a] # Überschriften stehen in pub-title
    m = re.search(r"year\s*=\s*(\d\d\d\d)", soup.find('textarea').text)
    if m:
        year.append(m.group(1))
    else:
        year.append("Ohne Jahr")
    m = re.search(r"title\s*=\s*{(.*?)}", soup.find('textarea').text)
    if m:
        title.append(m.group(1))
    else:
        title.append("Ohne Titel")
    m = re.search(r"author\s*=\s*{(.*?)}", soup.find('textarea').text)
    if m:
        author.append(m.group(1).replace(" and ", ",").split(","))
    else:
        author.append("Kein Autor")
    m = re.search(r"journal\s*=\s*{(.*?)}", soup.find('textarea').text)
    if m:
        journal.append(m.group(1))
    else:
        journal.append("Andere Veröffentlichungsform")

textfile = "\n".join(abstracts)
tokens = nltk.word_tokenize(textfile)
text = nltk.Text(tokens)
text_clean = [word for word in text if word not in ',-:()']

print "Done."
```

#### Wichtigste Autoren und Journals

```python
all_authors = [a for sublist in author for a in sublist]

print "Autoren"
pt = PrettyTable(field_names=['Autor', 'Publikationen']) 
c = Counter(all_authors)
[ pt.add_row(kv) for kv in c.most_common()[:20] ]
pt.align['Trigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```

```python
print "Journals"
pt = PrettyTable(field_names=['Journal', 'Publikationen']) 
c = Counter(journal)
[ pt.add_row(kv) for kv in c.most_common()[:20] ]
pt.align['Trigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```

```python
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()

# ------------------------------------------------------
print "Wichtigste Bigramme"
finder = BigramCollocationFinder.from_words(text_clean)
scored = finder.score_ngrams(bigram_measures.likelihood_ratio)

pt = PrettyTable(field_names=['Bigramm', 'Likelihood Ratio']) 
[ pt.add_row(kv) for kv in scored[:50] ]
pt.align['Bigramm'], pt.align['Likelihood Ratio'] = 'l', 'r' # Set column alignment
print pt

# ------------------------------------------------------
print "Häufigste Trigramme"
finder = TrigramCollocationFinder.from_words(text_clean)
scored = finder.score_ngrams(trigram_measures.raw_freq)

pt = PrettyTable(field_names=['Trigramm', 'Count']) 
[ pt.add_row(kv) for kv in scored[:20] ]
pt.align['Trigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```

```python
# ------------------------------------------------------
print "Häufigste Trigramme mit spannenden Wörtern"
finder = TrigramCollocationFinder.from_words(text_clean)
scored = finder.score_ngrams(trigram_measures.raw_freq)

pt = PrettyTable(field_names=['Trigramm', 'Count']) 
[ pt.add_row(kv) for kv in scored[:100] if (kv[0][2] not in stop)]
pt.align['Trigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt

# ------------------------------------------------------
print "Häufigste wichtige Bigramme"
finder = BigramCollocationFinder.from_words(text_clean)
scored = finder.score_ngrams(bigram_measures.raw_freq)
scored_no_stop = [bi for bi in scored if (bi[0][0] not in stop) and (bi[0][1] not in stop)]

pt = PrettyTable(field_names=['Bigramm', 'Count']) 
[ pt.add_row(kv) for kv in scored_no_stop[:50] ]
pt.align['Bigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```

#### Häufige Bigramme für einen Begriff

```python
search = "markov"

finder = BigramCollocationFinder.from_words(text_clean)
scored = finder.score_ngrams(bigram_measures.raw_freq)
scored_no_stop = [bi for bi in scored if (search in bi[0][0]) or (search in bi[0][1])]

pt = PrettyTable(field_names=['Bigramm', 'Count']) 
[ pt.add_row(kv) for kv in scored_no_stop[:50] ]
pt.align['Bigramm'], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```

#### Texte nach Inhalten clustern (Cosinus-Ähnlichkeit und K-Means)

```python
from nltk.cluster import KMeansClusterer, GAAClusterer, euclidean_distance
import nltk.stem

wnl = nltk.WordNetLemmatizer()
stopwords = set(nltk.corpus.stopwords.words('english'))

all_words = set(text_clean)

def normalize_word(word):
    return wnl.lemmatize(word.lower())

def vectorspaced(abstract_text):
    abstract_words = [normalize_word(word) for word in abstract_text.split()]
    return np.array([
        word in abstract_words and not word in stop
        for word in all_words], np.short)

clusterer = KMeansClusterer(5, euclidean_distance)
#clusterer = GAAClusterer(5)
clusterer.cluster([vectorspaced(a) for a in abstracts])
example = [clusterer.classify(vectorspaced(a)) for a in abstracts]

for id, t in sorted(zip(example, title)):
    print id, t
```
